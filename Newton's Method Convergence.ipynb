```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its derivative
def f(x):
    return x**2 - 2

def f_prime(x):
    return 2 * x

# Newton's method implementation
def newton_method(f, f_prime, x0, iterations):
    xs = [x0]
    for _ in range(iterations):
        x_next = xs[-1] - f(xs[-1]) / f_prime(xs[-1])
        xs.append(x_next)
    return xs

# Parameters
x0 = 1.0
iterations = 5

# Run Newton's method
xs = newton_method(f, f_prime, x0, iterations)

# Print results
print("Iteration |   x_n    |  f(x_n)  ")
for i, x in enumerate(xs):
    print(f"{i:9d} | {x:8.6f} | {f(x):9.2e}")

# Compute errors |f(x_n)|
errors = [abs(f(x)) for x in xs]

# Plot convergence (error vs iteration)
plt.figure()
plt.plot(list(range(len(errors))), errors)
plt.xlabel('Iteration')
plt.ylabel('Absolute error |f(xₙ)|')
plt.title("Newton's Method Convergence for f(x) = x² - 2")
plt.show()
```
